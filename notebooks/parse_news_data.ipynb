```python
import pandas as pd
import os
#load the data 
data_dir = 'D:/Learning DSFS/news_correlation_10ac_week0/data'
news_data = pd.read_csv(os.path.join(data_dir, 'rating.csv'))
domains_data = pd.read_csv(os.path.join(data_dir,'domains_location.csv'))
traffic_data = pd.read_csv(os.path.join(data_dir,'traffic.csv'))
news_data
```

```python
def get_top_10_websites(news_data):
    top_10_websites = news_data.groupby('source_name')['article_id'].count().nlargest(10)
    print(top_10_websites)
    return top_10_websites

def get_bottom_10_websites(news_data):
    bottom_10_websites = news_data.groupby('source_name')['article_id'].count().nsmallest(10)
    
    
    return bottom_10_websites

get_top_10_websites(news_data)
get_bottom_10_websites(news_data)
```

2 Websites with the hieghts number of visitors

```python
def get_top_bottom_websites_traffic(traffic_data):
    bottom_10_websites = traffic_data.nlargest(10, 'GlobalRank')
    top_10_websites = traffic_data.nsmallest(10, 'GlobalRank')
    return top_10_websites, bottom_10_websites

top_websites, bottom_websites = get_top_bottom_websites_traffic(traffic_data)
print("Top 10 Websites with Highest Traffic:")
print(top_websites[['Domain', 'GlobalRank']])

print("\nBottom 10 Websites with Lowest Traffic:")
print(bottom_websites[['Domain', 'GlobalRank']])
```

Top 10 Countries with the Highest Number of News Media Organizations:

```python


top_10_countries = domains_data.groupby('Country')['SourceCommonName'].count().nlargest(10)
bottom_10_countries = domains_data.groupby('Country')['SourceCommonName'].count().nsmallest(10)

print("Top 10 Countries with the Highest Number of News Media Organizations:")
print(top_10_countries)

print("\nBottom 10 Countries with the Lowest Number of News Media Organizations:")
print(bottom_10_countries)
```

1.3 Countries that have many articles written about them - the content of the news is about that country

```python

import pandas as pd
import spacy

# Load English tokenizer, tagger, parser, NER and word vectors
nlp = spacy.load("en_core_web_sm")

# Read the data
data = news_data

# Function to get countries from content
def get_countries(content):
    doc = nlp(content)
    return [ent.text for ent in doc.ents if ent.label_ == 'GPE']

# Apply the function to the 'content' column
data['countries'] = data['content'].apply(get_countries)

# Now you can count the occurrences of each country
country_counts = data['countries'].explode().value_counts()

# Print the top 10 countries
print(country_counts.nlargest(10))
```
